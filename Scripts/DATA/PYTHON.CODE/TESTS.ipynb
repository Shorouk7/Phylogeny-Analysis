```python
from Bio import SeqIO
import numpy as np

# Load aligned FASTA
fasta_file = "c:\\Users\\sh_de\\Desktop\\PHYLOGENTIC\\PHylogeny paper\\Phylogeny\\DATA\\cox.fas"  # your aligned FASTA file
records = list(SeqIO.parse(fasta_file, "fasta"))

# Convert to NumPy array of shape (n_samples, n_sites)
seq_array = np.array([list(record.seq) for record in records])
```

```python
import allel

# Make a numeric representation (A:0, C:1, G:2, T:3, others:-1)
def nucleotide_to_int(nuc):
    return {'A': 0, 'C': 1, 'G': 2, 'T': 3}.get(nuc.upper(), -1)

genotype_array = np.vectorize(nucleotide_to_int)(seq_array)

# Transpose to shape (n_variants, n_samples, ploidy=1)
g = allel.GenotypeArray(genotype_array.T[:, :, None])
```

```python
sample_ids = [
    "A. amoreuxi.A.AM1", "A. amoreuxi.A.AM2", "A. amoreuxi.M.AM1",
    "A. amoreuxi.M.AM1.2", "A. amoreuxi.M.AM2", "A. amoreuxi.T.AM2",
    "A. amoreuxi.T.AM2.2", "A. bicolor.E.B", "A. bicolor.T.B", "A. bicolor.T.B.2",
    "A.amoreuxi.Egy", "A.bicolor.Egy", "A.crassicauda.Egy", "B.arenicola.Egy",
    "L.quinquestriatus.Egy", "D.krausi"
]

# Define populations
pop_amoreuxi = [i for i, s in enumerate(sample_ids) if "amoreuxi" in s.lower()]
pop_bicolor = [i for i, s in enumerate(sample_ids) if "bicolor" in s.lower()]
pop_crassicauda = [i for i, s in enumerate(sample_ids) if "crassicauda" in s.lower()]
pop_arenicola = [i for i, s in enumerate(sample_ids) if "arenicola" in s.lower()]
pop_quinquestriatus = [i for i, s in enumerate(sample_ids) if "quinquestriatus" in s.lower()]
pop_krausi = [i for i, s in enumerate(sample_ids) if "krausi" in s.lower()]
```

```python
import numpy as np
import allel

# Assuming you already have your GenotypeArray g and your population lists:
# For example:
# pop_amoreuxi = [0, 1, 2, 3, 4, 5, 6]
# pop_bicolor = [7, 8, 9, 11]
# pop_other = [10, 12, 13, 14, 15]
subpops = [pop_amoreuxi, pop_bicolor, pop_crassicauda,pop_arenicola,pop_quinquestriatus , pop_krausi ]

# Calculate Weir & Cockerham's Fst
num, den = allel.weir_cockerham_fst(g, subpops=subpops)

# Mean Fst per pair
fst = np.sum(num, axis=1) / np.sum(den, axis=1)

# Display Fst values between each population and the rest
for i, value in enumerate(fst):
    print(f"FST for pop {i+1} vs others: {value:.4f}")

```
